# Импорт необходимых библиотек
Импортируйте стандартные библиотеки Python, такие как sys или os, для работы с кодом.

```python
def nth_term(a1, d, n):
    if n == 1:
        return a1
    return nth_term(a1, d, n - 1) + d


def sum_of_terms(a1, d, n):
    if n == 1:
        return a1
    return sum_of_terms(a1, d, n - 1) + nth_term(a1, d, n)


a1 = float(input("Введіть a1: "))
d = float(input("Введіть d: "))
n = int(input("Введіть n: "))

print("n-й член:", nth_term(a1, d, n))
print("Сума:", sum_of_terms(a1, d, n))
```

# Использование циклов и условий
Напишите код с циклами (for, while) и условными операторами (if-else) для демонстрации логики.
